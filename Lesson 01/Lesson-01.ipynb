# PIL short for pillow

PIL is a library used for working with images in pyhton. <br>
there are many libraries but we are using the below


```python
from PIL import Image

# to load the image we use the below code
image = Image.open("original_image.jpg")

# to display image
image.show(title="fun1")
```

```python
# other way to display image is to use the matplotlib library
import matplotlib.pyplot as plt

# to display image using matplotlib
plt.imshow(image)

# the attribute format is the extension of the image
print(image.format)

print(image.size)

print(image.mode)
```

# Operations on image

```python
# to perform operations on image.
from PIL import ImageOps

# converting image to grayscale
image_gray = ImageOps.grayscale(image)

# to display image in grayscale 
image_gray.show(title="fun1")

# to save image
image_gray.save("gray_scale.jpg")
```

###### Quantize: <br> In simple terms, quantization is a process of approximating or reducing the number of possible values in a set. It's like rounding off or grouping similar things together.


```python
# to quantize image
image_gray.quantize(2)

# split the red, blue and green portin in an image
red, green, blue = image.split()
red.show(title="red")
green.show(title="green")
blue.show(title="blue")
```

##### convert an image to an array

# cv2 short for opencv
is another library used to deal with images in the program

```python
# import numpy library
import numpy as np

# grayscale image array
print(np.array(image_gray))

# RGB color array
print("\nRGB array")
print(np.array(image))
```

```python
import cv2
```

```python
# load the image
image2 = cv2.imread("original_image.jpg")

# image is a numpyarray when we load it using cv2
print(type(image2))

# to print the image
import matplotlib.pyplot as plt
plt.imshow(image2)
plt.show()
```

```python
# the image in cv2 is loaded in BGR form so to convert it to 
# RGB we use the following code

new_image = cv2.cvtColor(image2, cv2.COLOR_BGR2RGB)
plt.imshow(new_image)
plt.show()
```

```python
# to convert image into gray scale we use
# python is case sensitive so remeber its "cvtColor" not "cvtcolor"

gray_scale_image = cv2.cvtColor(new_image, cv2.COLOR_RGB2GRAY)
plt.imshow(gray_scale_image)
plt.show()
```
